## Load and Preprocess Images

### Subtask:
Load the images from the extracted dataset, resize them to a uniform size (e.g., 64x64 pixels), and convert them into a numerical format suitable for machine learning.


**Reasoning**:
First, I will install the `opencv-python` library, which is essential for image processing tasks such as reading and resizing images, as specified in the subtask instructions. This ensures all required tools are available before proceeding.



````python
print("Installing OpenCV...")
!pip install opencv-python
print("OpenCV installed successfully.")
````

**Reasoning**:
Now that OpenCV is installed, I will proceed with loading and preprocessing the images as specified in the subtask. This involves importing necessary libraries, defining dataset paths, iterating through image directories, reading and resizing images using OpenCV, and assigning numerical labels, then converting them into NumPy arrays.



````python
import os
import cv2
import numpy as np

# Define target image size
IMG_SIZE = 64

# Define the base directory where the dataset was extracted
BASE_DIR = 'dogs_vs_cats/dataset/'
TRAIN_DIR = os.path.join(BASE_DIR, 'training_set')
TEST_DIR = os.path.join(BASE_DIR, 'test_set')

def load_and_preprocess_images(directory):
    images = []
    labels = []
    for category in ['cats', 'dogs']:
        path = os.path.join(directory, category)
        class_num = 0 if category == 'cats' else 1 # 0 for cats, 1 for dogs
        print(f"Processing {category} in {directory}...")
        for img_name in os.listdir(path):
            try:
                img_array = cv2.imread(os.path.join(path, img_name), cv2.IMREAD_COLOR)
                new_array = cv2.resize(img_array, (IMG_SIZE, IMG_SIZE))
                images.append(new_array)
                labels.append(class_num)
            except Exception as e:
                print(f"Error loading image {img_name}: {e}")
    return np.array(images), np.array(labels)

print("Loading and preprocessing training images...")
X_train, y_train = load_and_preprocess_images(TRAIN_DIR)
print("Training images loaded and preprocessed.")

print("Loading and preprocessing test images...")
X_test, y_test = load_and_preprocess_images(TEST_DIR)
print("Test images loaded and preprocessed.")

print(f"Shape of training features (X_train): {X_train.shape}")
print(f"Shape of training labels (y_train): {y_train.shape}")
print(f"Shape of test features (X_test): {X_test.shape}")
print(f"Shape of test labels (y_test): {y_test.shape}")
````

## Prepare Data for SVM

### Subtask:
Flatten the preprocessed image data into feature vectors (X) and normalize pixel values.


**Reasoning**:
I will flatten the image data `X_train` and `X_test` into 2D arrays and then normalize their pixel values to be between 0 and 1, as instructed in the subtask. This prepares the data for model training.



````python
print("Flattening and normalizing image data...")

# Flatten X_train and X_test
# Each image of shape (64, 64, 3) becomes a vector of length 64 * 64 * 3 = 12288
X_train_flattened = X_train.reshape(X_train.shape[0], -1)
X_test_flattened = X_test.reshape(X_test.shape[0], -1)

# Normalize pixel values to the range [0, 1]
X_train_normalized = X_train_flattened / 255.0
X_test_normalized = X_test_flattened / 255.0

print("Image data flattened and normalized.")
print(f"Shape of flattened and normalized training features (X_train_normalized): {X_train_normalized.shape}")
print(f"Shape of flattened and normalized test features (X_test_normalized): {X_test_normalized.shape}")
````

## Train SVM Classifier

### Subtask:
Initialize and train a Support Vector Machine (SVM) classifier from scikit-learn using the prepared training data.


**Reasoning**:
First, I will install the `scikit-learn` library, which is necessary for importing and using the Support Vector Machine (SVM) classifier as specified in the subtask instructions. This ensures the environment is set up correctly before proceeding with model training.

**Reasoning**:
Now that scikit-learn is installed, I will import the `SVC` class, instantiate the SVM classifier, and then train it using the `X_train_normalized` features and `y_train` labels, as outlined in the subtask instructions.



````python
from sklearn.svm import SVC

print("Initializing SVM classifier...")
# Initialize SVC classifier
# Using default parameters. verbose=True to monitor progress.
# Set random_state for reproducibility.
svm_classifier = SVC(random_state=42, verbose=True)

print("Training SVM classifier... This may take a while.")
# Train the SVM classifier
svm_classifier.fit(X_train_normalized, y_train)

print("SVM classifier training complete.")
````

## Evaluate Model Performance

### Subtask:
Evaluate the trained SVM model's performance on the test set. Calculate and display key metrics such as accuracy, precision, recall, and F1-score. A classification report and confusion matrix would also be beneficial.


**Reasoning**:
I will now evaluate the SVM model's performance on the test set by making predictions, calculating key metrics (accuracy, precision, recall, F1-score), and generating a confusion matrix and classification report as instructed in the subtask.



````python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, classification_report
import seaborn as sns
import matplotlib.pyplot as plt

print("Making predictions on the test set...")
y_pred = svm_classifier.predict(X_test_normalized)
print("Predictions complete.")

print("\n--- Model Performance Metrics ---")

# Calculate and print accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy:.4f}")

# Calculate and print precision (using 'weighted' for multiclass, 'binary' for binary)
precision = precision_score(y_test, y_pred, average='binary')
print(f"Precision: {precision:.4f}")

# Calculate and print recall (using 'weighted' for multiclass, 'binary' for binary)
recall = recall_score(y_test, y_pred, average='binary')
print(f"Recall: {recall:.4f}")

# Calculate and print F1-score (using 'weighted' for multiclass, 'binary' for binary)
f1 = f1_score(y_test, y_pred, average='binary')
print(f"F1-Score: {f1:.4f}")

print("\n--- Confusion Matrix ---")
# Generate and print the confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print(conf_matrix)

# Visualize confusion matrix
plt.figure(figsize=(6, 5))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Cat', 'Dog'], yticklabels=['Cat', 'Dog'])
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()

print("\n--- Classification Report ---")
# Generate and print the classification report
class_report = classification_report(y_test, y_pred, target_names=['Cat', 'Dog'])
print(class_report)
````
